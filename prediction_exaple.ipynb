```python
import matplotlib.pyplot as plt
import tensorflow as tf 

def drow_pictures(image_path, loaded_model):

    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.cast(image, tf.float32)  # Cast to tf.float32
    # image = tf.image.resize(image, [32, 32])
    image = image / 255.0
    image_gray = tf.image.rgb_to_grayscale(image)
    image_gray = tf.expand_dims(image_gray, axis=0)

    image_predicted = loaded_model(image_gray)

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(9, 4.5), tight_layout=True)

    ax1.imshow(image_gray[0], cmap='gray')
    ax2.imshow(image_predicted[0])
    ax3.imshow(image)

    ax2.set_title("Autoencoder results")

    ax1.set_xlabel('input')
    ax2.set_xlabel('output')
    ax3.set_xlabel('expected')
```

```python
from keras.models import load_model
from Autoencoder import Autoencoder, Encoder, Decoder
from main import ssim_loss
import cv2 
import tensorflow as tf 
import matplotlib.pyplot as plt
import numpy as np 
from tensorflow.keras.applications import VGG16

loaded_model = load_model("checkpoints/2.keras",
                          custom_objects={
        'Encoder': Encoder,
        'Decoder': Decoder,
        'Autoencoder': Autoencoder,
        'ssim_loss': ssim_loss
    }
)
```
